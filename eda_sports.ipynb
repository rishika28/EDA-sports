### Merging the data sets to get better insights on the information provided by them 

```python
season=data[['id','season','winner']]
merge=data2.merge(season,how='inner',left_on='match_id',right_on='id')
data.columns.values
```

```python
merge.info()
```

```python
merge.describe()
```

```python
data = data.drop(columns=["umpire3"],axis=1)
data.head()
```

### Analysis of wins per season for each team 

```python
wins=data.groupby("season")["winner"].value_counts()
wins
```

### Number of matches played in each season 

```python
plt.figure(figsize = (20,10))
sns.countplot('season',data=data,palette="dark")
plt.title("Number of Matches played in each IPL season",fontsize=20)
plt.xlabel("season",fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.show()
```

### Number of matches won by teams until now 

```python
plt.figure(figsize = (20,10))
sns.countplot(x='winner',data=data, palette='bright')
plt.title("Numbers of matches won by team ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("No of wins",fontsize=15)
plt.show()
```

### Toss results 

```python
Toss=data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['yellow', 'orange']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=20)
plt.show()
```

### Number of matches won according to the toss results 

```python
plt.figure(figsize = (20,10))
sns.countplot('season',hue='toss_decision',data=data,palette='dark')
plt.title("Numbers of matches won according to Toss result ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

### Number of matches won by batting or bowling first 

```python
data['win_by']=np.where(data['win_by_runs']>0,'Bat first','Bowl first')
plt.figure(figsize = (20,10))
sns.countplot('season',hue='win_by',data=data,palette='bright')
plt.title("Numbers of matches won by batting or bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

### Top Players 

```python
plt.figure(figsize = (20,10))
top_players = data.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="cool");
plt.show()
```

### Man of the match 

```python
final_matches=data.drop_duplicates(subset=['season'], keep='last')
final_matches[['winner','player_of_match']].reset_index(drop=True)
```

### Number of fours hit by players 

```python
four_data=merge[merge['batsman_runs']==4]
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='red')
plt.title("Numbers of fours hit by players ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

### Number of fours hit in each season 

```python
# we will plot graph on no of four hit in each season
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = 'blue')
plt.title("Numbers of fours hit in each season ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

### No. of sixes hit by players 

```python
six_data=merge[merge['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='green')
plt.title("Numbers of six hit by players ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of six",fontsize=15)
plt.show()
```

### Number of sixes in each season

```python
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = 'orange')
plt.title("Numbers of sixes hit in each season ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

### Top 10 leading scorers 

```python
batsman_score=data2.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

### Bowler who took the maximum wickets

```python
wicket_data=data2.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
